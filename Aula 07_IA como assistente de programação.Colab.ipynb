```
def calculadora(num1, num2, operador):
    if operador == '+':
        return num1 + num2
    elif operador == '-':
        return num1 - num2
    elif operador == '*':
        return num1 * num2
    elif operador == '/':
        if num2 == 0:
            return 0  # Resultado zero em caso de divisão por zero
        else:
            return num1 / num2
    else:
        return "Operador inválido"

# Para usar, basta chamar a função:
print(f"Teste 10 + 5: {calculadora(10, 5, '+')}")
print(f"Teste 40 / 0: {calculadora(40, 0, '/')}")
```

```
import random

def jogar_forca():
    # Lista de palavras focada em filmes
    palavras = ['matrix', 'inception', 'avatar', 'coringa', 'gladiador', 'interstellar']
    palavra_secreta = random.choice(palavras).lower()
    letras_corretas = ['_' for _ in palavra_secreta]
    tentativas = 6
    letras_tentadas = []

    print('--- JOGO DA FORCA: EDIÇÃO FILMES ---')

    while tentativas > 0 and '_' in letras_corretas:
        print(f'\nPalavra: {" ".join(letras_corretas)}')
        print(f'Tentativas restantes: {tentativas}')
        print(f'Letras já tentadas: {", ".join(letras_tentadas)}')

        chute = input('Digite uma letra: ').lower().strip()

        if not chute or len(chute) > 1:
            print('Por favor, digite apenas uma letra.')
            continue

        if chute in letras_tentadas:
            print('Você já tentou essa letra!')
            continue

        letras_tentadas.append(chute)

        if chute in palavra_secreta:
            for i, letra in enumerate(palavra_secreta):
                if letra == chute:
                    letras_corretas[i] = chute
            print(f'Boa! A letra "{chute}" existe na palavra.')
        else:
            tentativas -= 1
            print(f'Que pena! A letra "{chute}" não existe.')

    if '_' not in letras_corretas:
        print(f'\nParabéns! Você venceu! O filme era: {palavra_secreta}')
    else:
        print(f'\nGame Over! Você perdeu. O filme era: {palavra_secreta}')

# Chamando a função para testar agora:
jogar_forca()
```

```
import requests

def consultar_cep(cep):
    # Limpa o CEP (remove pontos e traços)
    cep = cep.replace("-", "").replace(".", "")

    # URL da API ViaCEP
    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        # Faz a requisição HTTP GET
        response = requests.get(url)

        # Verifica se a requisição foi bem sucedida
        if response.status_code == 200:
            dados = response.json()

            # Verifica se o CEP não foi encontrado
            if "erro" in dados:
                return "CEP não encontrado."

            return dados
        else:
            return f"Erro na requisição: {response.status_code}"

    except Exception as e:
        return f"Ocorreu um erro: {e}"

# Exemplo de uso
cep_teste = "01001-000"
resultado = consultar_cep(cep_teste)

print(f"--- Resultado da Consulta (CEP: {cep_teste}) ---")
print(resultado)
```

```
# Teste com o CEP da Avenida Paulista
cep_paulista = "01311-200"
resultado_paulista = consultar_cep(cep_paulista)

print(f"--- Resultado da Consulta (CEP: {cep_paulista}) ---")
print(resultado_paulista)
```
