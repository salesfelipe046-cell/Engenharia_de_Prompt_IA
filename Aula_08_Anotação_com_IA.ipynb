#Missão 03

```python
import requests

# Define uma função para consultar a API ViaCEP
def consultar_cep(cep):
    # Remove qualquer caractere não numérico do CEP
    cep = ''.join(filter(str.isdigit, cep))

    # Valida o formato do CEP
    if len(cep) != 8:
        return {"status": "erro", "mensagem": "CEP inválido. Deve conter 8 dígitos."}

    # Formata o CEP para a URL da API
    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        # Faz a requisição GET para a API
        response = requests.get(url)
        # Levanta um erro HTTP para códigos de status de erro (4xx ou 5xx)
        response.raise_for_status()

        # Converte a resposta JSON para um dicionário Python
        data = response.json()

        # Verifica se a resposta contém um erro (indicado por 'erro': true)
        if 'erro' in data and data['erro']:
            return {"status": "erro", "mensagem": "CEP não encontrado ou inválido."}
        else:
            return {"status": "sucesso", "dados": data}

    except requests.exceptions.HTTPError as http_err:
        return {"status": "erro", "mensagem": f"Erro HTTP: {http_err}"}
    except requests.exceptions.ConnectionError as conn_err:
        return {"status": "erro", "mensagem": f"Erro de Conexão: {conn_err}. Verifique sua conexão com a internet."}
    except requests.exceptions.Timeout as timeout_err:
        return {"status": "erro", "mensagem": f"Tempo limite da requisição excedido: {timeout_err}"}
    except requests.exceptions.RequestException as req_err:
        return {"status": "erro", "mensagem": f"Erro na requisição: {req_err}"}
    except ValueError:
        return {"status": "erro", "mensagem": "Resposta da API não é um JSON válido."}
```

```python
# Função para exibir os resultados de forma legível
def exibir_endereco(resultado):
    if resultado['status'] == 'sucesso':
        dados = resultado['dados']
        print("\n--- Informações do Endereço ---")
        print(f"CEP: {dados.get('cep', 'N/A')}")
        print(f"Logradouro: {dados.get('logradouro', 'N/A')}")
        print(f"Complemento: {dados.get('complemento', 'N/A')}")
        print(f"Bairro: {dados.get('bairro', 'N/A')}")
        print(f"Localidade: {dados.get('localidade', 'N/A')}")
        print(f"UF: {dados.get('uf', 'N/A')}")
        print(f"IBGE: {dados.get('ibge', 'N/A')}")
        print("------------------------------")
    else:
        print(f"\nErro: {resultado['mensagem']}")

# --- Exemplos de uso ---
print("##### Testando a função consultar_cep #####")

# Exemplo 1: CEP válido
cep_valido = "01001000"  # Exemplo: Praça da Sé, São Paulo
print(f"\nConsultando CEP válido: {cep_valido}")
resultado_valido = consultar_cep(cep_valido)
exibir_endereco(resultado_valido)

# Exemplo 2: CEP com formato inválido (menos de 8 dígitos)
cep_invalido_formato = "12345"
print(f"\nConsultando CEP com formato inválido: {cep_invalido_formato}")
resultado_invalido_formato = consultar_cep(cep_invalido_formato)
exibir_endereco(resultado_invalido_formato)

# Exemplo 3: CEP inexistente (que a API ViaCEP indica como erro)
cep_inexistente = "99999999"
print(f"\nConsultando CEP inexistente: {cep_inexistente}")
resultado_inexistente = consultar_cep(cep_inexistente)
exibir_endereco(resultado_inexistente)

# Exemplo 4: CEP com caracteres não numéricos
cep_com_traco = "05424-020" # Exemplo: Rua dos Pinheiros, São Paulo
print(f"\nConsultando CEP com caracteres não numéricos: {cep_com_traco}")
resultado_com_traco = consultar_cep(cep_com_traco)
exibir_endereco(resultado_com_traco)
```
