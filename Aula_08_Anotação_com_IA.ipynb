```python
'''
*Crie um script python simples para verificar a temperatura de um sistema. Se for maior que 30, exiba um alerta de superaquecimento.
'''

# Simula a temperatura atual do sistema. Você pode mudar este valor.
temperatura = 32

# Imprime a temperatura atual do sistema na tela.
print(f"Temperatura atual do sistema: {temperatura}°C")

# Verifica se a temperatura é maior que 30 graus usando uma condição 'if'.
if temperatura > 30:
    # Se a condição for verdadeira, exibe um alerta de superaquecimento.
    print("ALERTA: Superaquecimento detectado! A temperatura está acima de 30°C.")
else:
    # Se a condição for falsa (temperatura não é maior que 30), exibe uma mensagem de temperatura normal.
    print("Temperatura normal. O sistema está funcionando dentro dos limites.")
```
