```
def calcular_imc(peso, altura):
    if altura <= 0:
        return "Erro: A altura deve ser maior que zero."
    imc = peso / (altura ** 2)
    return imc

while True:
    try:
        idade = int(input("Digite sua idade: "))
        peso = float(input("Digite seu peso em kg: "))
        altura = float(input("Digite sua altura em metros: "))

        imc_calculado = calcular_imc(peso, altura)

        if isinstance(imc_calculado, str):
            print(imc_calculado)
        else:
            print(f"Sua idade é: {idade} anos")
            print(f"Seu IMC é: {imc_calculado:.2f}")

            if imc_calculado < 18.5:
                print("Classificação: Abaixo do peso")
            elif 18.5 <= imc_calculado < 24.9:
                print("Classificação: Peso normal")
            elif 25 <= imc_calculado < 29.9:
                print("Classificação: Sobrepeso")
            elif 30 <= imc_calculado < 34.9:
                print("Classificação: Obesidade Grau I")
            elif 35 <= imc_calculado < 39.9:
                print("Classificação: Obesidade Grau II")
            else:
                print("Classificação: Obesidade Grau III (Mórbida)")
    except ValueError:
        print("Entrada inválida. Por favor, digite números para idade, peso e altura.")

    continuar = input("Deseja calcular outro IMC? (s/n): ").lower()
    if continuar != 's':
        break

print("Obrigado por usar a calculadora de IMC!")
```

Diferente de uma versão simples, o programa foi aprimorado para solicitar também a idade do usuário, além de peso e altura. Essa adição torna o código mais completo e abre possibilidade para futuras melhorias, como análises mais detalhadas com base na faixa etária.

Outro ponto importante é a implementação de um loop (estrutura de repetição), que permite ao usuário realizar vários cálculos consecutivos sem precisar reiniciar o programa. Ao final de cada execução, o sistema pergunta se o usuário deseja continuar ou encerrar.
